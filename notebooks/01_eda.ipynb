```python
# Set up plotting style
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

# Create visualization
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# Bar chart
df['readmitted'].value_counts().plot(kind='bar', ax=ax[0], 
                                      color=['#2ecc71', '#e74c3c', '#95a5a6'])
ax[0].set_title('Readmission Distribution', fontsize=14, fontweight='bold')
ax[0].set_xlabel('Readmission Status')
ax[0].set_ylabel('Count')
ax[0].tick_params(axis='x', rotation=45)

# Pie chart
df['readmitted'].value_counts().plot(kind='pie', ax=ax[1], autopct='%1.1f%%',
                                     colors=['#2ecc71', '#e74c3c', '#95a5a6'])
ax[1].set_title('Readmission Percentage', fontsize=14, fontweight='bold')
ax[1].set_ylabel('')

plt.tight_layout()
plt.show()
```

```python
# Count missing values
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)

print("Columns with missing values:")
print(missing)

# Visualize missing data
if len(missing) > 0:
    plt.figure(figsize=(10, 6))
    missing.plot(kind='bar')
    plt.title('Missing Values by Column', fontsize=14, fontweight='bold')
    plt.xlabel('Number of Missing Values')
    plt.tight_layout()
    plt.show()
```

```python
# Age is in ranges like '[70-80)', '[50-60)', etc.
print("Age Distribution:")
print(df['age'].value_counts().sort_index())

# Visualize
plt.figure(figsize=(12, 6))
age_counts = df['age'].value_counts().sort_index()
age_counts.plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('Patient Age Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Age Group')
plt.ylabel('Number of Patients')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Does age affect readmission?
age_readmit = pd.crosstab(df['age'], df['readmitted'], normalize='index') * 100

# Plot
plt.figure(figsize=(14, 6))
age_readmit.plot(kind='bar', stacked=True, 
                 color=['#2ecc71', '#e74c3c', '#95a5a6'])
plt.title('Readmission Rate by Age Group', fontsize=14, fontweight='bold')
plt.xlabel('Age Group')
plt.ylabel('Percentage')
plt.legend(title='Readmitted', bbox_to_anchor=(1.05, 1))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Print the numbers
print("\nReadmission percentages by age:")
print(age_readmit)
```

```python
# Gender distribution
print("Gender Distribution:")
print(df['gender'].value_counts())

# Gender vs Readmission
gender_readmit = pd.crosstab(df['gender'], df['readmitted'], normalize='index') * 100

plt.figure(figsize=(10, 6))
gender_readmit.plot(kind='bar', color=['#2ecc71', '#e74c3c', '#95a5a6'])
plt.title('Readmission Rate by Gender', fontsize=14, fontweight='bold')
plt.xlabel('Gender')
plt.ylabel('Percentage')
plt.legend(title='Readmitted')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# How long do patients stay?
print("Time in Hospital Statistics:")
print(df['time_in_hospital'].describe())

# Visualize
plt.figure(figsize=(12, 6))
df['time_in_hospital'].hist(bins=14, edgecolor='black', color='coral')
plt.title('Distribution of Hospital Stay Duration', fontsize=14, fontweight='bold')
plt.xlabel('Days in Hospital')
plt.ylabel('Number of Patients')
plt.tight_layout()
plt.show()
```

```python
# Does longer stay = more readmission?
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='readmitted', y='time_in_hospital', palette='Set2')
plt.title('Hospital Stay Duration by Readmission Status', fontsize=14, fontweight='bold')
plt.xlabel('Readmission Status')
plt.ylabel('Days in Hospital')
plt.tight_layout()
plt.show()

# Statistical summary
print("\nAverage days by readmission status:")
print(df.groupby('readmitted')['time_in_hospital'].mean())
```

```python
# Medication statistics
print("Number of Medications Statistics:")
print(df['num_medications'].describe())

# Visualize
plt.figure(figsize=(12, 6))
df['num_medications'].hist(bins=30, edgecolor='black', color='lightgreen')
plt.title('Distribution of Number of Medications', fontsize=14, fontweight='bold')
plt.xlabel('Number of Medications')
plt.ylabel('Number of Patients')
plt.axvline(df['num_medications'].mean(), color='red', linestyle='--', 
            label=f'Mean: {df["num_medications"].mean():.1f}')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# More medications = more readmission?
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='readmitted', y='num_medications', palette='Set3')
plt.title('Number of Medications by Readmission Status', fontsize=14, fontweight='bold')
plt.xlabel('Readmission Status')
plt.ylabel('Number of Medications')
plt.tight_layout()
plt.show()
```

```python
# Check previous emergency visits
print("Previous Emergency Visits:")
print(df['number_emergency'].value_counts().sort_index().head(10))

# Previous inpatient visits
print("\nPrevious Inpatient Visits:")
print(df['number_inpatient'].value_counts().sort_index().head(10))

# Visualize
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# Emergency visits
emergency_counts = df['number_emergency'].value_counts().sort_index().head(10)
emergency_counts.plot(kind='bar', ax=ax[0], color='salmon', edgecolor='black')
ax[0].set_title('Previous Emergency Visits', fontsize=12, fontweight='bold')
ax[0].set_xlabel('Number of Visits')
ax[0].set_ylabel('Number of Patients')

# Inpatient visits
inpatient_counts = df['number_inpatient'].value_counts().sort_index().head(10)
inpatient_counts.plot(kind='bar', ax=ax[1], color='skyblue', edgecolor='black')
ax[1].set_title('Previous Inpatient Visits', fontsize=12, fontweight='bold')
ax[1].set_xlabel('Number of Visits')
ax[1].set_ylabel('Number of Patients')

plt.tight_layout()
plt.show()
```

```python
# This is usually a STRONG predictor!

# Emergency visits vs readmission
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='readmitted', y='number_emergency', palette='Reds')
plt.title('Previous Emergency Visits by Readmission Status', fontsize=14, fontweight='bold')
plt.xlabel('Readmission Status')
plt.ylabel('Number of Emergency Visits')
plt.tight_layout()
plt.show()

# Inpatient visits vs readmission
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='readmitted', y='number_inpatient', palette='Blues')
plt.title('Previous Inpatient Visits by Readmission Status', fontsize=14, fontweight='bold')
plt.xlabel('Readmission Status')
plt.ylabel('Number of Inpatient Visits')
plt.tight_layout()
plt.show()

# Print averages
print("\nAverage previous emergency visits by readmission:")
print(df.groupby('readmitted')['number_emergency'].mean())

print("\nAverage previous inpatient visits by readmission:")
print(df.groupby('readmitted')['number_inpatient'].mean())
```

```python
# Create a 2x2 dashboard of key insights
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Plot 1: Age vs Readmission
age_readmit = pd.crosstab(df['age'], df['readmitted'], normalize='index') * 100
age_readmit.plot(kind='bar', stacked=True, ax=axes[0,0], 
                 color=['#2ecc71', '#e74c3c', '#95a5a6'])
axes[0,0].set_title('Readmission Rate by Age Group', fontsize=12, fontweight='bold')
axes[0,0].set_xlabel('Age Group')
axes[0,0].set_ylabel('Percentage')
axes[0,0].legend(title='Readmitted', loc='upper left', fontsize=8)
axes[0,0].tick_params(axis='x', rotation=45, labelsize=8)

# Plot 2: Time in Hospital
sns.boxplot(data=df, x='readmitted', y='time_in_hospital', ax=axes[0,1], palette='Set2')
axes[0,1].set_title('Hospital Stay by Readmission', fontsize=12, fontweight='bold')
axes[0,1].set_xlabel('Readmission Status')
axes[0,1].set_ylabel('Days in Hospital')

# Plot 3: Number of Medications
sns.boxplot(data=df, x='readmitted', y='num_medications', ax=axes[1,0], palette='Set3')
axes[1,0].set_title('Medications by Readmission', fontsize=12, fontweight='bold')
axes[1,0].set_xlabel('Readmission Status')
axes[1,0].set_ylabel('Number of Medications')

# Plot 4: Previous Inpatient Visits
sns.boxplot(data=df, x='readmitted', y='number_inpatient', ax=axes[1,1], palette='Blues')
axes[1,1].set_title('Previous Hospitalizations by Readmission', fontsize=12, fontweight='bold')
axes[1,1].set_xlabel('Readmission Status')
axes[1,1].set_ylabel('Previous Inpatient Visits')

plt.tight_layout()
plt.savefig('../visualizations/eda_dashboard.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Dashboard saved to visualizations/eda_dashboard.png")
```

## 🔍 Key Findings from EDA

### 1. Readmission Patterns
- **11.2%** of patients are readmitted within 30 days
- **35%** of patients have some form of readmission
- This is a class imbalance problem - we'll need to handle it!

### 2. Age Insights
- Readmission rates increase with age
- Highest risk: 70-80 age group
- Elderly patients need more intensive discharge planning

### 3. Strongest Predictors (Initial Observations)
**Previous hospital utilization is KEY:**
- Patients with prior inpatient visits: MUCH higher readmission risk
- Patients with prior emergency visits: Higher readmission risk
- Previous history is more important than demographics!

**Clinical complexity matters:**
- More medications → Higher readmission risk
- Longer hospital stays → Higher readmission risk (sicker patients)
- More diagnoses → Higher readmission risk

### 4. Medication Management
- Average: 16 medications per patient (poly-pharmacy!)
- Insulin is very common (diabetes population)
- Medication changes during stay may impact outcomes

### 5. Next Steps for Feature Engineering
- Create binary target: `<30 days readmission` = 1, else = 0
- Engineer features from previous utilization history
- Create medication complexity indicators
- Handle missing values strategically
- Encode categorical variables properly

### 6. Business Impact Opportunity
- If we can identify high-risk patients, we can:
  - Provide intensive discharge planning
  - Schedule early follow-up appointments
  - Arrange home health services
  - Medication reconciliation
  - Reduce costly readmissions!
